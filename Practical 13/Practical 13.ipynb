```python
import pandas as pd

from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import StandardScaler, MinMaxScaler
```

```python
# -------------------------------
# 📌 Step 1: Load Dataset
# -------------------------------
df = pd.read_csv('transformation_dataset.csv')

print("Original Data:\n")
print(df.head())
```

```python
# -------------------------------
# ✅ Step 1: Label Encoding
# -------------------------------
le = LabelEncoder()
df['Department'] = le.fit_transform(df['Department'])

# -------------------------------
# ✅ Step 2: One-Hot Encoding
# -------------------------------
df = pd.get_dummies(df, columns=['Gender'], drop_first=True)

print("\nAfter Encoding:\n")
print(df.head())
```

```python
scaler = StandardScaler()
df[['Age', 'Salary']] = scaler.fit_transform(df[['Age', 'Salary']])
print("Standard Scaling:-\n",df)


scaler = MinMaxScaler()
df[['Age', 'Salary']] = scaler.fit_transform(df[['Age', 'Salary']])
print("\nMin-Max Scaling:-\n",df)
```

```python
# Select only relevant columns
df_subset = df[['Age', 'Salary', 'Department', 'Gender_Male']]
print(df_subset.head())
```

```python
# Group data by Department
dept_groups = df.groupby('Department')
for dept, group in dept_groups:
    print(f"\nDepartment: {dept}\n")
print(group.head())
```
